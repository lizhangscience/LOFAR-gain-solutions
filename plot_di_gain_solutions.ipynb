# NCP field, observation number L205861

1st sagecal step: Direction-independent calibration, column 1.

Sagecal solution interval is ~10 seconds. Here, data are plotted every 5 minutes, i.e., every 30th data point is plotted.

Gain metric, $G$ = âˆšTrace(JJ<sup>H</sup>)

```python
import numpy as np
import pyfits
import time
import os
import matplotlib.pyplot as plt
import scipy.signal as signal

def load_gain_solutions(datafile):

    #gain_solutions is a dictionary of the form [cluster id][station index][frequency index][time index][jones term]
    t1 = time.time()
    gain_solutions = np.load(datafile)['data'].item()
    num_stations = np.load(datafile)['num_stations'].item()    
    freq_range = np.load(datafile)['freq_range']
    time_range = np.load(datafile)['time_range']
    cluster_id = np.load(datafile)['cluster_id'].item()
    t2 = time.time()
    load_time = (t2-t1)/60

    return (gain_solutions, num_stations, freq_range, time_range, cluster_id, load_time)

def compute_jones_matrix_element(gain_solutions, num_stations, freq_range, time_range, cluster_id, matrix_element):
    #Compute element of Jones matrix. Choices can be 'XX', 'XY', 'YX', 'YY'
    matrix_element_correspondence = {'XX':['S0', 'S1'], 'XY':['S2', 'S3'], 'YX':['S4', 'S5'], 'YY':['S6', 'S7']}            
    gain = []
    
    for station_num in range(num_stations):

        gain_station = []
        for i in range(len(freq_range)):
             gain_freq = []
             for k in range(len(time_range)):
                 s0 = gain_solutions[cluster_id][station_num][i][k][matrix_element_correspondence[matrix_element][0]]
                 s1 = gain_solutions[cluster_id][station_num][i][k][matrix_element_correspondence[matrix_element][1]]
                 gain_time = s0 + s1*1j
                 gain_freq.append(gain_time)
             gain_station.append(gain_freq)
        gain.append(gain_station)    
    gain = np.array(gain)

    return gain

def compute_jones_matrix(gain_solutions, num_stations, freq_range, time_range, cluster_id):
    
    matrix_element_correspondence = {'XX':['S0', 'S1'], 'XY':['S2', 'S3'], 'YX':['S4', 'S5'], 'YY':['S6', 'S7']}           
    gain = []
    
    for station_num in range(num_stations):
        gain_station = []
        for i in range(len(freq_range)):
             gain_freq = []
             for k in range(len(time_range)):
                 gain_element = gain_solutions[cluster_id][station_num][i][k]
                 g_xx = gain_element['S0'] + 1j*gain_element['S1']
                 g_xy = gain_element['S2'] + 1j*gain_element['S3']
                 g_yx = gain_element['S4'] + 1j*gain_element['S5']
                 g_yy = gain_element['S6'] + 1j*gain_element['S7']      
                 g = np.matrix([[g_xx, g_xy], [g_yx, g_yy]])
                 gain_time = g
                 gain_freq.append(gain_time)
             gain_station.append(gain_freq)
        gain.append(gain_station)    
    gain = np.array(gain)
    return gain

def compute_jones_matrix_eigenvalues(gain_solutions, num_stations, freq_range, time_range, cluster_id):
    
    matrix_element_correspondence = {'XX':['S0', 'S1'], 'XY':['S2', 'S3'], 'YX':['S4', 'S5'], 'YY':['S6', 'S7']}            
    gain = []
    
    for station_num in range(num_stations):
        gain_station = []
        for i in range(len(freq_range)):
             gain_freq = []
             for k in range(len(time_range)):
                 gain_element = gain_solutions[cluster_id][station_num][i][k]
                 g_xx = gain_element['S0'] + 1j*gain_element['S1']
                 g_xy = gain_element['S2'] + 1j*gain_element['S3']
                 g_yx = gain_element['S4'] + 1j*gain_element['S5']
                 g_yy = gain_element['S6'] + 1j*gain_element['S7']             
                 T = g_xx + g_yy #Trace
                 D = g_xx*g_yy - g_xy*g_yx #Determinant
                 gain_time = [T/2 + (T**2/4 - D)**0.5, T/2 - (T**2/4 - D)**0.5]
                 gain_freq.append(gain_time)
             gain_station.append(gain_freq)
        gain.append(gain_station)    
    gain = np.array(gain)
    return gain
                                     
def compute_jones_matrix_norm(gain_solutions, num_stations, freq_range, time_range, cluster_id):
    
    matrix_element_correspondence = {'XX':['S0', 'S1'], 'XY':['S2', 'S3'], 'YX':['S4', 'S5'], 'YY':['S6', 'S7']}
            
    gain = []
    for station_num in range(num_stations):
        gain_station = []
        for i in range(len(freq_range)):
             gain_freq = []
             for k in range(len(time_range)):
                 gain_element = gain_solutions[cluster_id][station_num][i][k]
                 g_xx = gain_element['S0'] + 1j*gain_element['S1']
                 g_xy = gain_element['S2'] + 1j*gain_element['S3']
                 g_yx = gain_element['S4'] + 1j*gain_element['S5']
                 g_yy = gain_element['S6'] + 1j*gain_element['S7']      
                 g = np.matrix([[g_xx, g_xy], [g_yx, g_yy]])
                 gain_time = np.sqrt(np.trace(g*g.H))
                 gain_freq.append(gain_time)
             gain_station.append(gain_freq)
        gain.append(gain_station)    
    gain = np.array(gain)
    return gain
                                     
def compute_power_spectrum(gain, num_stations, time_range, freq_range):

    # Compute rate range
    n_time = gain[0,0].size
    timestep = time_range[1]-time_range[0]
    FT_rate = np.fft.fftfreq(n_time, d=timestep)
    rate_range = np.fft.fftshift(FT_rate)

    # Compute delay range
    freqstep = freq_range[1]-freq_range[0]
    n_freq =  int(np.round((freq_range[-1]-freq_range[0])/freqstep))
    FT_delay = np.fft.fftfreq(n_freq, d=freqstep)
    delay_range = np.fft.fftshift(FT_delay)
    delay_range=list(delay_range)
    delay_range.remove(0) #lombscargle complains if 0 is included
    delay_range=np.array(delay_range)

    #Used to normalize periodogram
    normval = len(freq_range)

    # List to hold sqrt(power spectrum), per station 
    PS = []
    gain0 = np.copy(gain)
    
    for i in range(gain.shape[0]):
        gain0[i] = gain[i] - np.mean(gain[i])

    for i in range(num_stations):

        #Perform Fourier Transform wrt time. Input gains are complex.
        FT_wrt_time = []
        for j in range(len(freq_range)):
            FT = np.fft.fft(gain0[i,j], axis=0)
            FT = np.fft.fftshift(FT)
            FT_wrt_time.append(FT)
        #FT_wrt_time = np.array(np.abs(FT_wrt_time))
        FT_wrt_time = np.array(FT_wrt_time)
        #Compute periodogram wrt frequency
        pgram_wrt_freq = [] 
        for j in range(len(time_range)):
            #Compute periodogram for real and imaginary parts separately, and them to compute the power spectrum
            pgram = signal.lombscargle(freq_range, FT_wrt_time[:,j].real, delay_range) + signal.lombscargle(freq_range, FT_wrt_time[:,j].imag, delay_range) 
            #Normalized periodogram
            pgram_wrt_freq.append((np.sqrt(4*(pgram/normval))))
            #pgram_wrt_freq.append(pgram)
        PS_station = pgram_wrt_freq
        PS_station = np.swapaxes(PS_station,0,1) #Swap time and frequency axes
        PS.append(PS_station)
        
    power_spectrum = np.array(PS)

    return (power_spectrum, rate_range, delay_range)

def read_station_positions(filename):
    station_positions = {}
    station_num = 0
    with open(filename) as f:
        for l in f:
            line = l.split(' ')
            if line[0]!='#':
                if line[0][:2]=='CS' and line[0][-4:]=='HBA0':
                    coods = {}
                    coods['X'] = float(line[1])
                    coods['Y'] = float(line[2])
                    coods['Z'] = float(line[3])
                    station_positions[station_num] = coods
                    station_num = station_num+1
                if line[0][:2]=='CS' and line[0][-4:]=='HBA1':
                    coods = {}
                    coods['X'] = float(line[1])
                    coods['Y'] = float(line[2])
                    coods['Z'] = float(line[3])
                    station_positions[station_num] = coods
                    station_num = station_num+1
                if line[0][:2]=='RS':
                    coods = {}
                    coods['X'] = float(line[1])
                    coods['Y'] = float(line[2])
                    coods['Z'] = float(line[3])
                    station_positions[station_num] = coods
                    station_num = station_num+1
                
    f.close()
   
    return station_positions    

def station_id_number_correspondence(filename):
#Find station number corresponding to a station ID
    station_correspondence = {}
    station_num = 0
    with open(filename) as f:
        for l in f:
            line = l.split(' ')
            if line[0]!='#':
                if line[0][:2]=='CS' and line[0][-4:]=='HBA0':
                    station_correspondence[line[0]] = station_num 
                    station_num = station_num+1
                if line[0][:2]=='CS' and line[0][-4:]=='HBA1':
                    station_correspondence[line[0]] = station_num
                    station_num = station_num+1
                if line[0][:2]=='RS' and line[0][-3:]=='HBA':
                    station_correspondence[line[0]] = station_num
                    station_num = station_num+1
                
    f.close()
       
    return station_correspondence

def compute_distance(station1, station2):

    station_positions = read_station_positions('LOFAR_station_positions.txt')

    x1 = station_positions[station1]['X']
    y1 = station_positions[station1]['Y']
    z1 = station_positions[station1]['Z']

    x2 = station_positions[station2]['X']
    y2 = station_positions[station2]['Y']
    z2 = station_positions[station2]['Z']
          
    distance = np.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)

    return distance

def compute_uv_distance(station1, station2, freq_MHz):

    station_positions = read_station_positions('LOFAR_station_positions.txt')

    x1 = station_positions[station1]['X']
    y1 = station_positions[station1]['Y']
    z1 = station_positions[station1]['Z']

    x2 = station_positions[station2]['X']
    y2 = station_positions[station2]['Y']
    z2 = station_positions[station2]['Z']

    wavelength = 3.e2/freq_MHz          

    baseline_length = np.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)
    uv_distance = baseline_length/wavelength

    return uv_distance

def abs_gain_product_vs_uv_distance(gain, num_stations, frequency_index, freq_range):  
    gain_prod = [] 
    distance = []       
 
    for station1 in range(num_stations):
        for station2 in range(station1+1, num_stations):
            distance.append(compute_uv_distance(station1, station2, freq_range[frequency_index]))
            gain_prod_time = []
            for k in range(len(time_range)):
                    gain1 = gain[station1][frequency_index][k]
                    gain2 = gain[station2][frequency_index][k]
                    gain_prod_time.append(np.abs(gain1)*np.abs(gain2))
            gain_prod.append(gain_prod_time)   
  
    sort_index = np.argsort(distance)
    uv_distance = [distance[i] for i in sort_index]
    gain_product = [gain_prod[i] for i in sort_index]

    gain_product = np.array(gain_product)

    mean_gain_product = [np.mean(np.array(gain_product[i])) for i in range(len(gain_product))]
    variance_gain_product = [np.var(np.array(gain_product[i])) for i in range(len(gain_product))]

    return (mean_gain_product, variance_gain_product, uv_distance)

def abs_gain_jones_matrix_product_vs_uv_distance(gain, num_stations, frequency_index, freq_range):  
    gain_prod = [] 
    distance = []       
 
    for station1 in range(num_stations):
        for station2 in range(station1+1, num_stations):
            distance.append(compute_uv_distance(station1, station2, freq_range[frequency_index]))
            gain_prod_time = []
            for k in range(len(time_range)):
                    gain1 = np.matrix(gain[station1][frequency_index][k])
                    gain2 = np.matrix(gain[station2][frequency_index][k])
                    gain_prod_time.append(np.sqrt(np.abs(np.trace(gain1*gain2.H))))
            gain_prod.append(gain_prod_time)   
  
    sort_index = np.argsort(distance)
    uv_distance = [distance[i] for i in sort_index]
    gain_product = [gain_prod[i] for i in sort_index]

    gain_product = np.array(gain_product)

    mean_gain_product = [np.mean(np.array(gain_product[i])) for i in range(len(gain_product))]
    variance_gain_product = [np.var(np.array(gain_product[i])) for i in range(len(gain_product))]

    return (mean_gain_product, variance_gain_product, uv_distance)


def abs_gain_product_vs_uv_distance_one_station(gain, num_stations, station, frequency_index, freq_range):  
    gain_prod = [] 
    distance = []       
 
    station1 = station
    
    for station2 in range(num_stations):
            distance.append(compute_uv_distance(station1, station2, freq_range[frequency_index]))
            gain_prod_time = []
            for k in range(len(time_range)):
                    gain1 = gain[station1][frequency_index][k]
                    gain2 = gain[station2][frequency_index][k]
                    gain_prod_time.append(np.abs(gain1)*np.abs(gain2))
            gain_prod.append(gain_prod_time)   
  
    sort_index = np.argsort(distance)
    uv_distance = [distance[i] for i in sort_index]
    gain_product = [gain_prod[i] for i in sort_index]

    gain_product = np.array(gain_product)

    mean_gain_product = [np.mean(np.array(gain_product[i])) for i in range(len(gain_product))]
    variance_gain_product = [np.var(np.array(gain_product[i])) for i in range(len(gain_product))]

    return (mean_gain_product, variance_gain_product, uv_distance)

def abs_gain_jones_matrix_product_vs_uv_distance_one_station(gain, num_stations, station, frequency_index, freq_range):  
    gain_prod = [] 
    distance = []       
 
    station1 = station
    
    for station2 in range(num_stations):
            distance.append(compute_uv_distance(station1, station2, freq_range[frequency_index]))
            gain_prod_time = []
            for k in range(len(time_range)):
                    gain1 = np.matrix(gain[station1][frequency_index][k])
                    gain2 = np.matrix(gain[station2][frequency_index][k])
                    gain_prod_time.append(np.sqrt(np.abs(np.trace(gain1*gain2.H))))
            gain_prod.append(gain_prod_time)   
  
    sort_index = np.argsort(distance)
    uv_distance = [distance[i] for i in sort_index]
    gain_product = [gain_prod[i] for i in sort_index]

    gain_product = np.array(gain_product)

    mean_gain_product = [np.mean(np.array(gain_product[i])) for i in range(len(gain_product))]
    variance_gain_product = [np.var(np.array(gain_product[i])) for i in range(len(gain_product))]

    return (mean_gain_product, variance_gain_product, uv_distance)

def plot_data (x, y, plot_format='o', xlabel=None, ylabel=None, title=None, filename=None):
    plt.plot(x, y, '%s'%plot_format)
    if xlabel:
        plt.xlabel('%s'%xlabel)
    if ylabel:
        plt.ylabel('%s'%ylabel)
    if title:
        plt.title('%s'%title)
    if filename:
        plt.savefig('%s'%filename)
        plt.close()
    else:
        plt.show()

def plot_data_semilogx (x, y, plot_format='o', xlabel=None, ylabel=None, title=None, filename=None):
    plt.semilogx(x, y, '%s'%plot_format)
    if xlabel:
        plt.xlabel('%s'%xlabel)
    if ylabel:
        plt.ylabel('%s'%ylabel)
    if title:
        plt.title('%s'%title)
    if filename:
        plt.savefig('%s'%filename)
        plt.close()
    else:
        plt.show()
        
def plot_data_image(img):
    plt.imshow(img)
    plt.colorbar()
```
